# ПРЕДОБРАБОТКА ДАННЫХ

# 1. Обработка пропусков

Пропущенные значения в датасете отсутствуют, дополнительная обработка не требуется.

# 2. Обработка выбросов

Выбросы присутствуют в следующих категориях: Slepp_Duration, Study_Hours, Social_Media_hours, Stress_level.

Обнаруженные выбросы не являются явно аномальными и могут отражать реальные значения, поэтому на данном этапе они не удаляются.

# 3. Кодирование категорий

Категориальные признаки: Gender, Department.
Преобразуем их в числовой формат с использованием LabelEncoder.

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender'] = le.fit_transform(df['Gender'])
df['Department'] = le.fit_transform(df['Department'])
````

# 4. Разделение данных

Для обучения и оценки моделей данные будут разделены на обучающую , тестовую и валидационную выборки с использованием функции train_test_split из библиотеки sklearn.

Доля тестовой выборки составила 20% от общего объёма данных, валидационной — 20%, обучающей - 60%.

````python
from sklearn.model_selection import train_test_split

# Регрессия (CGPA) (успеваемость)
X_reg = df.drop(columns=['CGPA'])
y_reg = df['CGPA']

X_train_reg, X_temp_reg, y_train_reg, y_temp_reg = train_test_split(
    X_reg, y_reg, test_size=0.4, random_state=44
)

X_val_reg, X_test_reg, y_val_reg, y_test_reg = train_test_split(
    X_temp_reg, y_temp_reg, test_size=0.5, random_state=44
)

# Классификация (Depression) (уровень депрессии)
X_clf = df.drop(columns=['Depression'])
y_clf = df['Depression']

X_train_clf, X_temp_clf, y_train_clf, y_temp_clf = train_test_split(
    X_clf, y_clf, test_size=0.4, random_state=44
)

X_val_clf, X_test_clf, y_val_clf, y_test_clf = train_test_split(
    X_temp_clf, y_temp_clf, test_size=0.5, random_state=44
)
````

На данном этапе, наш датасет разделен на три категории: обучающая, валидационная и тестовая части.

# Масштабирование данных

Масштабирование признаков необходимо для приведения их к единому масштабу, что предотвращает доминирование признаков с большими числовыми значениями и улучшает работу моделей машинного обучения. Масштабирование будет выполнено с использованием StandardScaler класса.

````python
# Регрессия 

from sklearn.preprocessing import StandardScaler

scaler_reg = StandardScaler()

# обучаем scaler на train
X_train_reg = scaler_reg.fit_transform(X_train_reg)

# применяем к val и test
X_val_reg = scaler_reg.transform(X_val_reg)
X_test_reg = scaler_reg.transform(X_test_reg)
````

````python
# Классификация 

scaler_clf = StandardScaler()

# обучаем scaler на train
X_train_clf = scaler_clf.fit_transform(X_train_clf)

# применяем к val и test
X_val_clf = scaler_clf.transform(X_val_clf)
X_test_clf = scaler_clf.transform(X_test_clf)
````

Масштабирование признаков было выполнено с использованием StandardScaler.

Параметры масштабирования были рассчитаны на обучающей выборке и затем применены к валидационной и тестовой выборкам.

# Обучение моделей

Сначала обучим модель линейной регрессии, затем модель логистической регрессии. В качестве метрик для модели линейной регрессии возьмем MAE, R^2, а в качестве метрик для модели логистической регрессии возьмем precision, recall из модуля classification_report.

# Модель линейной регрессии

````python
# Обучаем модель предсказывать CGPA (успеваемость) студента на основе его режима занятий и сна

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train_reg, y_train_reg)
````

````python
# делаем предсказания на валидационной выборке

y_pred_val = model.predict(X_val_reg)

# Оцениваем качество модели на валидационной выборке
from sklearn.metrics import mean_absolute_error, r2_score

print("VAL MAE:", mean_absolute_error(y_val_reg, y_pred_val))
print("VAL R2:", r2_score(y_val_reg, y_pred_val))


print("\n" + 30*"=" + "\n")

# Делаем предсказания на тестовой выборке

y_pred_test = model.predict(X_test_reg)

# финальная оценка модели
print("TEST MAE:", mean_absolute_error(y_test_reg, y_pred_test))
print("TEST R2:", r2_score(y_test_reg, y_pred_test))
````

Модель линейной регрессии показала низкое качество предсказания целевой переменной CGPA. Значение коэффициента детерминации R² оказалось близким к нулю (≈ 0.05), что свидетельствует о слабой зависимости между признаками и целевой переменной. Средняя абсолютная ошибка (MAE ≈ 0.45) указывает на заметную погрешность предсказаний. При этом близость значений метрик на валидационной и тестовой выборках говорит об отсутствии переобучения модели.

# Модель логистической регресии

````python
# Обучаем модель предсказывать риск развития депрессии студента исходя из его образа жизни и уровня стресса

from sklearn.linear_model  import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(X_train_clf, y_train_clf)
````

````python
# делаем предсказания на валидационной выборке

y_pred_val = model.predict(X_val_clf)

# Оцениваем качество модели на валидационной выборке

from sklearn.metrics import classification_report

print(classification_report(y_val_clf, y_pred_val))


print("\n" + 30*"=" + "\n")

# Делаем предсказания на тестовой выборке

y_pred_test = model.predict(X_test_clf)

# финальная оценка модели
print(classification_report(y_test_clf, y_pred_test))


````

Модель логистической регрессии показала высокое значение accuracy (~0.90), однако данный результат является некорректным из-за сильного дисбаланса классов. Анализ метрик precision и recall показал, что модель не распознаёт класс 1 (Depression), так как значения данных метрик равны нулю. Это означает, что модель фактически предсказывает только доминирующий класс и не решает поставленную задачу классификации.

ДОБАВИМ НЕМНОГО ЭКСПЕРИМЕНТОВ И ПОИГРАЕМСЯ С КОЛИЧЕСТВОМ ЭПОХ И ШАГОМ ОБУЧЕНИЯ)))

````python
# логистическая регрессия, обучаемая градиентным спуском
from sklearn.linear_model import SGDClassifier
for epoch in [100, 500, 1000, 2000]:
    print(f"Quantity of epochs = {epoch}")
    for lr in [0.0001, 0.001, 0.01, 0.1]:
        print(f"Learning rate = {lr}")

        model = SGDClassifier(loss='log_loss', learning_rate='constant', eta0=lr, max_iter=epoch)
        # loss - делает модель логистической регрессией, learning rate - шаг обучения не меняется во время обучения, 
        # eta0 - значение шага обучения, max_iter - количество эпох (сколько раз модель проходит по данным)

        model.fit(X_train_clf, y_train_clf)

        y_pred_val = model.predict(X_val_clf)

        print(classification_report(y_val_clf, y_pred_val))
````

В ходе эксперимента с различными значениями learning rate и числа эпох было установлено, что данные параметры оказывают влияние на поведение модели.

При увеличении learning rate модель начинает чаще предсказывать класс 1 (Depression), что приводит к росту значений precision и recall.

Однако даже при оптимальных параметрах значения recall остаются низкими (не превышают 0.09), что свидетельствует о слабой способности модели выявлять случаи депрессии.
Увеличение числа эпох не приводит к существенному улучшению качества модели.

Таким образом, проведённый эксперимент показал, что изменение параметров обучения не решает проблему качества модели, которая обусловлена сильным дисбалансом классов в данных.

ПОКАЖЕМ ОТСУТСТВИЕ ПЕРЕОБУЧЕНИЯ ДЛЯ ЛОГИСТИЧЕСКОЙ РЕГРЕССИИ: 

````python
print("Train accuracy:", model.score(X_train_clf, y_train_clf))
print("Validation accuracy:", model.score(X_val_clf, y_val_clf))
print("Test accuracy:", model.score(X_test_clf, y_test_clf))
````

Высокие значения accuracy (~0.90) объясняются сильным дисбалансом классов, при котором доминирующий класс (Depression = 0) составляет около 90% выборки.

Модель фактически предсказывает только данный класс, что приводит к высокой точности, но не отражает реального качества классификации.

Близость значений accuracy на обучающей, валидационной и тестовой выборках свидетельствует об отсутствии переобучения, однако модель не справляется с распознаванием редкого первого класса.
